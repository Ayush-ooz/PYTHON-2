### Pandas Series
- pandas series is like a column in a table 
- it is a 1D array holding data of any type

```python
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(myvar)
print(type(myvar))
print(myvar[1])
```

```python
# for customize index
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a,index=['x','y','z'])
print(myvar)
print(myvar[1])
```

```python
a=[1.4,7,2]
myvar=pd.Series(a)
print(myvar)
```

```python
import pandas as pd
student_name=['ram','shyam','seeta','geeta']
marks=[60,70,80,90]
myvar=pd.Series(marks,index=student_name,name='student_result')
print(myvar)
```

```python
calories={'day1':400,'day2':380,'day3':480,'day4':370}
myvar=pd.Series(calories,name="Calories burn per day")
print(myvar)
```

```python
calories={'day1':400,'day2':380,'day3':480,'day4':370}
myvar=pd.Series(calories,name="Calories burn per day",index=['day1','day2'])
print(myvar)
```

### Series Attributes

```python
calories={'day1':400,'day2':380,'day3':480,'day4':370}
myvar=pd.Series(calories,name="Calories burn per day")
print(myvar.size)
print(myvar.dtype)
print(myvar.name)
print(myvar.index)
print(myvar.values)
print(myvar.is_unique)
print(myvar)
```

## DataFrame
- DataFrame is a 2 dimensional data structure like a 2D Array or table with rows and columns

```python
data={"calories":[420,380,400,370],"duration":[50,40,45,35]}
df=pd.DataFrame(data)
df
```

```python
print(df.loc[0])
```

```python
print(df['calories'])
```

```python
print(df.loc[[0,1]])
```

```python
data={"calories":[420,380,400,370],"duration":[50,40,45,35]}
df=pd.DataFrame(data,index=['day1','day2','day3','day4'])
print(df)
print("--------------------------------------")
print(df.loc['day1'])
print(df.iloc['day1'])
```

```python
print(df[::2])
```

```python
data={"Name":["Alice","Bob","Charlie"],"Age":[25,30,35],"City":["New York","Los Angeles","Chicago"]}
df=pd.DataFrame(data)
print(df)
print("--------------------------------------")
print(df.loc[0:1])
print("--------------------------------------")
print(df.iloc[0:1])
print("--------------------------------------")
print(df.loc[(df.Age>26)])
```

```python
data={'A':[1,2,3,4],'B':[5,6,7,8]}
df=pd.DataFrame(data,index=['r1','r2','r3','r4'])
df
```
