```python
import pandas as pd
df1=pd.DataFrame({
    'Student_ID':[1,2,3],
    'Name':['Amit','Riya','John']
})
df2=pd.DataFrame({
    '1D':[1,2,4],
    'Marks':[85,90,75]
})
result=pd.merge(df1,df2,left_on="Student_ID",right_on="1D")
print(result)
```

```python
import pandas as pd
df1=pd.DataFrame({
    'Student_ID':[1,2,3],
    'Name':['Amit','Riya','John']
})
df2=pd.DataFrame({
    '1D':[1,2,4],
    'Marks':[85,90,75]
})
result=pd.merge(df1,df2,left_on="Student_ID",right_on="1D",how="outer")
print(result)
```

## groupby()
- used to split a DataFrame into groups based on one or more columns,allowing for efficient data analysis and aggregation

```python
data={'Product Category':['Electronics','Furniture','Electronics','Furniture'],
     'Sales':[200,150,300,100]}
df=pd.DataFrame(data)
df
```

```python
result=df.groupby('Product Category')['Sales'].sum()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].mean()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].median()
print(result)
```

```python
result=df.groupby('Product Category',as_index=True).count()#default value True
print(result)
```

```python
result=df.groupby('Product Category',as_index=False).count()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].describe()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].sample()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].size()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].count()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].nth(0)
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].nth(1)
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].groups
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].nunique()
print(result)
```

```python
df=pd.DataFrame({
    'Department':['Sales','Sales','HR','HR','IT','IT'],
    'Salary':[50000,60000,40000,45000,70000,75000],
    'Bonus':[5000,6000,2000,2500,8000,8500]
})
result=df.groupby('Department').agg({'Salary':['mean','max'],'Bonus':'sum'})
print(result)
```

```python
result=df.groupby('Department').agg('mean')
print(result)
```

```python
result=df.groupby('Department').agg(['mean','min','max'])
print(result)
```

```python
result=df.groupby('Department').agg(MeanSalary=('Salary','mean'),MaxSalary=('Salary','max'),TotalBonus=('Bonus','sum'))
print(result)
```
