## 1.Linear Regression
- statiscal method used to model the relationship between a dependent variable and one or more independent variables
- The Goal of LR is to find the best fitting line that describe the relationship between the variable

```python
import pandas as pd
df=pd.read_csv("test.csv")
df.info()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
print(type(x))
print(type(y))
```

```python
print(x)
print(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
print(ypred)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
diff=pd.DataFrame({"Actual":y_test,"Predicted":ypred})
diff
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
ypred=lr.predict(x)
print(ypred)
```

```python
print("coefficient:",lr.coef_)
print("intercept:",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y,ypred))
print("MSE:",metrics.mean_squared_error(y,ypred))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
ypred=lr.predict([[18]])
print(ypred)
```

```python
print("R2 Score:",metrics.r2_score(y,ypred))
```
