```python
import pandas as pd
df=pd.DataFrame({"Hours_Studied":[1,2,3,4,5,6,7],"Exam_Score":[50,55,65,70,75,80,85]})
df
```

```python
import numpy as np
x=np.array([[1],[2],[3],[4],[5],[6],[7]])
y=np.array([50,55,65,70,75,80,85])
print(x.shape)
print(y.shape)
```

```python
x=df.iloc[:,0:1]
# or x=df[["Hours_Studied"]]
y=df.iloc[:,-1]
# or y=df["Exam_Score"]
print(x)
print(y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
ypred=lr.predict(x)
diff=pd.DataFrame({"Actual":y,"Predicted":ypred})
diff
```

```python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y,ypred))
print("MSE:",metrics.mean_squared_error(y,ypred))
print("R2 Score:",metrics.r2_score(y,ypred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color="blue",label="Actual_Scores")
plt.plot(x,ypred,color="red",label="Predicted_Score")
plt.xlabel("Hours_Studied")
plt.ylabel("Exam_Score")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df=pd.get_dummies(data=df,drop_first=True)
df
```

```python
y=df["Selling_Price"]
x=df.iloc[:,1:]
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
diff=pd.DataFrame({"Actual":y_test,"Predicted":ypred})
diff
```

```python
print("coefficient:",lr.coef_)
print("intercept:",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y_test,ypred))
print("MSE:",metrics.mean_squared_error(y_test,ypred))
print("R2 Score:",metrics.r2_score(y_test,ypred))
```
